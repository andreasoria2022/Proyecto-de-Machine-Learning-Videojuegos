```python
def UserForGenre(genres):
    genero = df_merged[df_merged['genres'] == genres]

    if len(genero) == 0:
        return 'El género no existe'

    horas_por_año = []
    for i in range(len(genero)):
        año = genero.iloc[i]['año']
        horas = genero.iloc[i]['horas']
        horas_por_año.append({'Año': año, 'Horas': horas})

    user_id = genero['user_id'].values[0]

    return {"El usuario con más horas jugadas para género": genres, 'es' :user_id, 'Horas jugadas': horas_por_año}
```

```python
def UserForGenre(genres: str):
    # Filtrar los usuarios del año deseado
    genero = df_merged[df_merged['genres'] == genres]
    
    if len(genero) == 0:
        return 'El género no existe'

    horas_por_año = genero.apply(lambda x: {'Año': x['año'], 'Horas jugadas:': x['horas']}, axis=1).tolist()

    user_id = genero['user_id'].values[0]

    return {'El usuario con más horas jugadas para el género': genres, 'es' :user_id, 'Horas jugadas': horas_por_año}
```
